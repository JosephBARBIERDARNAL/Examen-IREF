```python
# Create a CSV file in memory
csv_file = io.StringIO()
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Question", "Option 1", "Option 2", "Option 3", "Réponse"])

# Add the questions and their options to the CSV file
for question in qcm_questions:
    csv_writer.writerow([question["question"], *question["options"], question["answer"]])

# Save the file
csv_file.seek(0)
output_path = 'risk.csv'
with open(output_path, 'w', encoding='utf-8') as f:
    f.write(csv_file.getvalue())
```
